```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
train_data = pd.read_csv('/content/Letter-TrainTest.csv')
predict_data = pd.read_csv('/content/Letter-Predict.csv')
# Chuẩn bị dữ liệu
label_column = 'class'

# Chuẩn bị dữ liệu
X_train = train_data.drop(label_column, axis=1)
y_train = train_data[label_column]
X_predict = predict_data.drop(label_column, axis=1)
y_predict = predict_data[label_column]




model_nb = GaussianNB()
model_nb.fit(X_train, y_train)

# Dự đoán và đánh giá
y_pred_nb = model_nb.predict(X_predict)
print("Kết quả Naive Bayes:")
print("Ma trận sai lầm:\n", confusion_matrix(y_predict, y_pred_nb))
print("Độ chính xác:", accuracy_score(y_predict, y_pred_nb))
print("Báo cáo phân loại:\n", classification_report(y_predict, y_pred_nb))
```

```python
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier

train_data = pd.read_csv('Letter-TrainTest.csv')
X_train = train_data.drop('class', axis=1)
y_train = train_data['class']

# Xác định phạm vi tham số để thử nghiệm
k_values = [3, 5, 7, 9, 11]  # Số lượng láng giềng
distance_metrics = ['euclidean', 'manhattan', 'minkowski']  # Độ đo khoảng cách



results = []


for k in k_values:
    for metric in distance_metrics:
        # Tạo mô hình KNN
        model = KNeighborsClassifier(n_neighbors=k, metric=metric)


        # Thực hiện K-fold cross-validation với K=10
        cv = KFold(n_splits=10, shuffle=True, random_state=42)  # random_state để đảm bảo tính tái lập
        scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')


        results.append({
            'k': k,
            'metric': metric,

            'mean_accuracy': scores.mean(),
            'std_accuracy': scores.std()
        })


for result in results:
    print(result)

# Chọn 3 mô hình tốt nhất dựa trên độ chính xác trung bình
top_3_models = sorted(results, key=lambda x: x['mean_accuracy'], reverse=True)[:3]
print("\n3 mô hình tốt nhất:")
for model in top_3_models:
    print(model)
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report

from sklearn.naive_bayes import GaussianNB # This line is added to import GaussianNB

def get_metrics(model, X_train, y_train, X_predict, y_predict):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_predict)
    accuracy = accuracy_score(y_predict, y_pred)
    report = classification_report(y_predict, y_pred, output_dict=True)
    precision = report['weighted avg']['precision']
    recall = report['weighted avg']['recall']
    f1_score = report['weighted avg']['f1-score']
    return accuracy, precision, recall, f1_score

# Get metrics for Naive Bayes
nb_accuracy, nb_precision, nb_recall, nb_f1 = get_metrics(GaussianNB(), X_train, y_train, X_predict, y_predict) # Assuming you imported GaussianNB

# Get metrics for top 3 KNN models
knn_metrics = []
for model_params in top_3_models:
    model = KNeighborsClassifier(n_neighbors=model_params['k'], metric=model_params['metric'])
    accuracy, precision, recall, f1_score = get_metrics(model, X_train, y_train, X_predict, y_predict)
    knn_metrics.append([accuracy, precision, recall, f1_score])

# Create the evaluation table
data = {
    'Mô hình': ['Naive Bayes', 'KNN (tham số 1)', 'KNN (tham số 2)', 'KNN (tham số 3)'],
    'Độ chính xác': [nb_accuracy] + [m[0] for m in knn_metrics],
    'Precision (trung bình)': [nb_precision] + [m[1] for m in knn_metrics],
    'Recall (trung bình)': [nb_recall] + [m[2] for m in knn_metrics],
    'F1-score (trung bình)': [nb_f1] + [m[3] for m in knn_metrics],
}

df = pd.DataFrame(data)
print(df.to_string(index=False))
```
