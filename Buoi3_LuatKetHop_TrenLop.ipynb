BÀI 2 : #không phải tổ hợp mà phải liệt kê tập hợp như Bài 1  (Sai bài 2 dãy só truyền và

```python
import itertools
#không phải tổ hợp mà phải liệt kê tập hợp như Bài 1
mon_hoc = ['CSLT', 'KTLT', 'OOP', 'CTDL', 'LTCSDL']
transactions_moi = []

# Tạo tất cả các tổ hợp có độ dài từ 1 đến 5
for i in range(1, len(mon_hoc) + 1):
  for to_hop in itertools.combinations(mon_hoc, i):
    transactions_moi.append(list(to_hop))

# In ra transactions_moi để kiểm tra
print(transactions_moi)
```

```python
!pip install pandas mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd # Import the pandas library and assign it to the alias 'pd'

te = TransactionEncoder()
te_ary = te.fit(transactions_moi).transform(transactions_moi)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(df, min_support=0.25, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
```

```python
# To see the generated rules, run the code.
rules
```
