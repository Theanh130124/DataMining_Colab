```python
import pandas as pd
```

```python
data = pd.read_excel("/content/Bai1.xlsx")
age_min = data['Age 1'].min()
age_max = data['Age 1'].max()
data['Age 1'] = round(((data['Age 1'] - age_min) / (age_max - age_min)) * 5, 1)
data
```

```python
data.loc[data['PerformanceRating'] == 1, ['PerformanceRating','PerformanceResult']] = [2, "Meet Expectation"]
```

```python
data.loc[data['PerformanceRating'] == 5, ['PerformanceRating','PerformanceResult']] = [4, "Exceed Expectation"]
```

```python
data
```

```python
EducationLevel = {
    "Below College ": 1,      # Below College -> 0
    "College ": 2,   # College -> 0.25
    " Bachelor": 3,    # Bachelor -> 0.5
    "Master": 4,   # Master -> 0.75
    "Doctor": 5       # Doctor -> 1
}

G = {
    "Male" : 0,
    "Female" : 1
}

Marital = {
    "Single" : 1,
    "Married" : 3,
    "Divorced" : 5
}

Age_2 = {
    "Young labor group": 1,
    "Main labor group": 2,
    "Middle-aged labor group": 3,
    "Old labor group":4

}

EducationBackground = {
    "Exceed Expectation" : 5,
    "Marketing" : 4,
    "Life Sciences" : 3,
    "Human Resources" : 2 ,
    "Medical" :  1,
    "Other" : 0
}

JobRole = {
    "Sales Executive" : 0,
    "Manager" : 1,
    "Developer" :2,
    "Sales Representative" : 3,
    "Human Resources" : 4,
    "Senior Developer" :5,
    "Data Scientist" :6
}

EnvironmentSatisfaction = {
    "Low" : 1,
    "Medium" :2,
    "High" : 3,
    "Very High" : 4
}

RelationshipSatisfaction = {
    "Low" : 1,
    "Medium" :2,
    "High" : 3,
    "Very High" : 4
}

WorkLifeBalance = {
    "Bad" : 1,
    "Good" :2,
    "Better" : 3,
    "Best" : 4
}

JobInvolvement = {
    "Low" : 1,
    "Medium" :2,
    "High" : 3,
    "Very High" : 4
}

TotalWorkExperienceInYears = {
    "no exps" : 1,
    "Copper range": 2,
    "Silver range" : 3,
    "Gold range" : 4,
    "Diamond range" : 5
}

ExperienceYearsInCurrentRole = {
    "no exps" : 1,
    "Copper range": 2,
    "Silver range" : 3,
    "Gold range" : 4,
    "Diamond range" : 5
}






data['Age 2'] = data['Age 2'].map(Age_2)
# data['Gender'] = data['Gender'].map(G)
# data['MaritalStatus'] = data['Marital'].map(Marital)
data['EducationBackground'] = data['EducationBackground'].map(EducationBackground)
data['JobRole'] = data['JobRole'].map(JobRole)



```

```python
data['Gender'] = data['Gender'].map(G)
data['MaritalStatus'] = data['MaritalStatus'].map(Marital)
```

```python
data.drop("Number", axis= 1, inplace= True)
```

```python
data.fillna(3, inplace= True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

trainSet , testSet = train_test_split(data, test_size = 0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(trainSet[["Age 1","Age 2",	"Gender",	"MaritalStatus", 	"EducationLevel",	"EducationBackground",	"JobRole",	"EnvironmentSatisfaction",
                  "RelationshipSatisfaction",	"WorkLifeBalance",	"TotalWorkExperienceInYears",	"ExperienceYearsInCurrentRole",	"PerformanceRating"]], trainSet["PerformanceResult"])

knn_pred = knn.predict(testSet[["Age 1","Age 2",	"Gender",	"MaritalStatus", 	"EducationLevel",	"EducationBackground",	"JobRole",	"EnvironmentSatisfaction",
                  "RelationshipSatisfaction",	"WorkLifeBalance",	"TotalWorkExperienceInYears",	"ExperienceYearsInCurrentRole",	"PerformanceRating"]])

print("Pan tich")
print(classification_report(testSet["PerformanceResult"], knn_pred))
```
