```python
import pandas as pd
file = ('/content/TNTHPT2023.csv')
df = pd.read_csv(file)
df.columns= ['SBD','Toán','Ngữ Văn','Ngoại Ngữ','Vật Lí','Hóa Học','Sinh Học','Lịch Sử','Địa lí','GCCD','Mã Ngoại Ngữ']
df.info()
```

```python
#Kiểm tra dữ liệu trùng lặp
duplicate_rows = df[df.duplicated(subset='SBD')]
print('Số lượng dữ liệu trùng lâp: ' ,len(duplicate_rows))
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer  # Import SimpleImputer

# Đọc dữ liệu từ file Excel
df = pd.read_csv('/content/thi_sinh_hcm.csv')

# Chọn các cột cần thiết và tạo DataFrame mới
df_cluster = df[['SBD', 'Toán', 'Ngữ Văn', 'Ngoại Ngữ']]

# Tính điểm tổ hợp (tổng 3 môn thành phần) - Thay đổi tên cột nếu cần
df_cluster['Tổ hợp'] = df[['Vật Lí', 'Hóa Học', 'Sinh Học']].sum(axis=1, skipna=True)

# Khởi tạo SimpleImputer để thay thế NaN bằng giá trị trung bình
imputer = SimpleImputer(strategy='mean')

# Chuẩn hóa dữ liệu trước khi gom cụm
# Áp dụng imputer trước khi chuẩn hóa
data_to_scale = df_cluster[['Toán', 'Ngữ Văn', 'Ngoại Ngữ', 'Tổ hợp']]
imputed_data = imputer.fit_transform(data_to_scale)  # Thay thế NaN

scaler = StandardScaler()
scaled_data = scaler.fit_transform(imputed_data)  # Chuẩn hóa dữ liệu đã được xử lý NaN


# Hàm để gom cụm và in kết quả
def cluster_and_analyze(k):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # n_init=10 for better convergence
    clusters = kmeans.fit_predict(scaled_data)
    df_cluster['Cụm'] = clusters

    print(f"\nKết quả gom cụm với K = {k}:")
    for i in range(k):
        cluster_data = df_cluster[df_cluster['Cụm'] == i]
        num_elements = len(cluster_data)
        percentage = (num_elements / len(df_cluster)) * 100
        print(f"  Cụm {i + 1}: {num_elements} phần tử, chiếm {percentage:.2f}%")

        # Phân tích đặc điểm của cụm (ví dụ: điểm trung bình của các môn)
        print(f"    Đặc điểm: Điểm trung bình Toán: {cluster_data['Toán'].mean():.2f}, "
              f"Ngữ Văn: {cluster_data['Ngữ Văn'].mean():.2f}, "
              f"Ngoại Ngữ: {cluster_data['Ngoại Ngữ'].mean():.2f}, "
              f"Tổ hợp: {cluster_data['Tổ hợp'].mean():.2f}")

# Gom cụm với K = 3, 5, 10
cluster_and_analyze(3)
cluster_and_analyze(5)
cluster_and_analyze(10)
```
