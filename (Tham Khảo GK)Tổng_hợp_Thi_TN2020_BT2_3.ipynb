Load dữ liệu lên

```python
import pandas as pd
file = ('/content/Full_Mark_2020.csv')
df = pd.read_csv(file)
df.columns= ['No','Code','Diali','GDCD','Hoahoc','KHTN','KHXH','LichSu','Ngoaingu','NguVan','Sinhhoc','Toan','Vatli','city']
df.info()

#No ra 1051860
#COde cũng vậy
```

Hai dòng đầu tiên chuyển đổi cột 'Vatli' và 'city' sang kiểu dữ liệu số. Nếu có lỗi, nó sẽ thay thế giá trị lỗi bằng NaN. df.info() sau đó được gọi để hiển thị thông tin tóm tắt về DataFrame df sau khi đã chuyển đổi kiểu dữ liệu của cột 'Vatli' và 'city'.

```python
#Chuyển dữ liệu từ object về dạng số
df['Code'] = pd.to_numeric(df['Code'], errors='coerce')
df['Vatli'] =pd.to_numeric(df['Vatli'], errors='coerce')
df['city'] = pd.to_numeric(df['city'], errors='coerce')

df.info()
```

```python
#Kiểm tra dữ liệu trùng lặp theo cột Code -> vì mã code là Unique
duplicate_rows = df[df.duplicated(subset='Code')]
print('Số lượng dữ liệu trùng lâp: ' ,len(duplicate_rows))
```

```python
df = df.drop_duplicates(subset='Code')
df
#860965 row x 14 col
```

```python
#Từ kết quả trên ta thấy có 971038 rows -> thí sinh -> copy ra một df y hệt
my_df = df
my_df['Code'] = range(1 , len(my_df) + 1)
my_df
```

```python
#Đếm số lượng thí sinh không thi cả hai mon KHTN và KHXH
count_no_exam = len(my_df[(my_df['KHTN'].isnull()) & (my_df['KHXH'].isnull())])
#Đếm số lượng thí sinh tham gia cả hai môn KHTN và KHXH
count_both_exam = len(my_df[(my_df['KHTN'].notnull()) & (my_df['KHXH'].notnull())])
#Hiển thị kết quả
print(f"Số lượng thí sinh không thi cả hai môn KHTN và KHXH: {count_no_exam}")
print(f'Số lượng thí sinh thi cả hai môn KHTN và KHXH: {count_both_exam}')
```

```python
#Tách dataframe my_df thành my_df_khtn và my_df_khxh
my_df_khtn = my_df[my_df['KHTN'].notnull()]
my_df_khxh = my_df[my_df['KHXH'].notnull()]
my_df_khtn.info()
my_df_khxh.info()
```

```python
#Lọc ra các cột cần tính toán của tất cả
my_df_khtn = my_df_khtn[['Toan','NguVan','Ngoaingu','Vatli','Hoahoc','Sinhhoc']]
my_df_khtn.info()
my_df_khxh = my_df_khxh[['Toan','NguVan','Ngoaingu','LichSu','Diali','GDCD']]
my_df_khxh.info()

#283779
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
#Tiền xử lý dữ liệu
my_df_khtn.fillna(my_df_khtn.mean(), inplace=True)
# my_df_khxh.fillna(my_df_khxh.mean(), inplace=True)
```

```python
#Tìm số cụm tối ưu sử dụng phương pháp Elbow
max_clusters = 10
sse = []
for k in range(2, max_clusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(my_df_khtn)
    sse.append(kmeans.inertia_)
#Vẽ biểu đồ Elbow
plt.plot(range(2, max_clusters + 1), sse, marker='o')
plt.xlabel('Số lượng cụm')
plt.ylabel('SSE')
plt.title('Phương pháp Elbow')
plt.show()
```

```python
#Số cụm tối ưu dưa trên biểu đồ Elbow
num_clusters = 5
#Ap dụng thuật toán K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(my_df_khtn)
#Gán nhãn cụm cho dữ liệu
my_df_khtn['Cluster'] = kmeans.labels_
#Vẽ biểu đồ phân cụm
plt.scatter(my_df_khtn['Cluster'],my_df_khtn['Toan'], c=my_df_khtn['Cluster'], cmap='viridis')
plt.xlabel('Cluster')
plt.ylabel('Toan')
plt.title('Phân cụm K-means')
plt.show()
#
```

```python
import numpy as np
def kmeans_display(X, label):
    K = np.amax(label) + 1
    X0 = X[label == 0, :]
    X1 = X[label == 1, :]
    X2 = X[label == 2, :]

    plt.plot(X0[:, 0], X0[:, 1], 'b^', markersize = 4, alpha = .8)
    plt.plot(X1[:, 0], X1[:, 1], 'go', markersize = 4, alpha = .8)
    plt.plot(X2[:, 0], X2[:, 1], 'rs', markersize = 4, alpha = .8)

    plt.axis('equal')
    plt.plot()
    plt.show()
  #Số cụm tối ưu dựa trên biểu đồ Elbow
num_clusters = 5
#Ap dụng thuật toán K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(my_df_khtn[['Toan','NguVan']])
#
print('Center found by sckiit-leanr:\n', kmeans.cluster_centers_)

pred_label = kmeans.predict(my_df_khtn[['Toan','NguVan']])
kmeans_display(my_df_khtn[['Toan','NguVan']].values, pred_label)
#
  #Gán nhãn cụm cho dữ liệu

```
